## MULTI INDEX PIVOT TABLE

```python
table1 = pd.pivot_table(df, index=['gender', 'heart_disease', 'smoking_status'])
table1
```

```python
import matplotlib.pyplot as plt

# Create a pivot table with 'gender', 'heart_disease', and 'smoking_status' as the indexes and 'age' as the values
table1 = pd.pivot_table(df, index=['gender', 'heart_disease', 'smoking_status'], values='age')

# Create a bar chart for each gender and smoking status combination, with the height of the bars representing the average age for that group
fig, ax = plt.subplots(figsize=(12,6))
for gender in ['Male', 'Female']:
    for smoking in ['never smoked', 'formerly smoked', 'smokes']:
        try:
            subset = table1.loc[(gender, 1, smoking),:]
            ax.bar(f"{gender} - {smoking}", subset.values[0], label=f"{gender} - {smoking}")
        except KeyError:
            pass

# Add a line chart showing the average age for each gender and heart disease combination
for gender in ['Male', 'Female']:
    try:
        subset = table1.loc[(gender, 1, slice(None)),:].mean(level=['gender'])
        ax.plot([f"{gender} - Heart Disease", f"{gender} - No Heart Disease"], subset.values, label=f"{gender} - Overall")
    except KeyError:
        pass
    
# Set the chart title and axis labels
ax.set_title("Average Age by Gender, Heart Disease, and Smoking Status")
ax.set_xlabel("Gender, Heart Disease, and Smoking Status")
ax.set_ylabel("Average Age")

# Add a legend
ax.legend()

# Display the chart
plt.show()
```

```python
# Create a pivot table with 'work_type' and 'gender' as the indexes and 'age' as the values
table2 = pd.pivot_table(df, index=['work_type', 'gender'], values='age')

# Print the pivot table
print(table2)
```

```python
# Create a pivot table with 'work_type' and 'gender' as the indexes and 'age' as the values
table2 = pd.pivot_table(df, index=['work_type', 'gender'], values='age')

# Plot a stacked bar chart
table2.unstack().plot(kind='bar', stacked=True)

# Add axis labels and title
plt.xlabel('Work Type')
plt.ylabel('Age')
plt.title('Age by Work Type and Gender')
plt.show()
```

## COLUMN LEVEL GROUPING

```python
# Create a pivot table with 'work_type' as the index, 'gender' as the columns, and 'age' as the values
table3 = pd.pivot_table(df, index='work_type', columns='gender', values='age')

# Print the pivot table
print(table3)
```

```python
# Create the bar chart
fig, ax = plt.subplots()
table3.plot(kind='bar', ax=ax)

# Set the axis labels and title
ax.set_xlabel('Work Type')
ax.set_ylabel('Age')
ax.set_title('Age by Work Type and Gender')

# Show the plot
plt.show()
```

## VALUES PIVOT TABLE

```python
# Create a pivot table with 'work_type' and 'gender' as the indexes, 'age' as the values, and 'hypertension' as the columns
table4 = pd.pivot_table(df, index=['work_type', 'gender'], columns='hypertension', values='age')

# Print the pivot table
print(table4)
```

```python
import matplotlib.pyplot as plt

# Create a stacked bar chart
table4.plot(kind='bar', stacked=True)

# Add axis labels and a title
plt.xlabel('Work Type and Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Work Type and Gender with Hypertension Status')

# Show the plot
plt.show()
```

## STYLING PIVOT TABLES

```python
# Create a pivot table with 'work_type' and 'gender' as the indexes and 'age' as the values
table = pd.pivot_table(df, index=['work_type', 'gender'], values='age')

# Style the pivot table
styled_table = table.style\
    .background_gradient(cmap='Blues', subset=['age'])\
    .highlight_max(subset=['age'], color='green')\
    .highlight_min(subset=['age'], color='red')

# Display the styled pivot table
display(styled_table)
```

```python
pivot = np.round(pd.pivot_table(df, values=['hypertension', 'age'], 
                                index=['work_type'], 
                                aggfunc=np.mean,
                                fill_value=0),2)
                                
pivot = pivot.reindex(pivot['age'].sort_values(ascending=False).index).nlargest(10, 'age')

pivot.style.format({'age':'${0:,.0f}',
                   'hypertension':'{0:,.0f}hp'}).bar(color='#d65f5f')
```
